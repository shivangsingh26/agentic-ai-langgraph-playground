```python
#Embedding using Huggingface

import os
from dotenv import load_dotenv
load_dotenv() # Load environment variables from .env file
```

```python
os.environ['HUGGINGFACE_ACCESS_TOKEN'] = os.getenv("HUGGINGFACE_ACCESS_TOKEN")
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
hf_embeddings_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2",
    #sentence-transformer is a SOTA model for sentence embeddings BY Huggingface
)
```

```python
text = "ArithmeticError is the base class for all errors that occur for numeric calculations."
embeddings = hf_embeddings_model.embed_query(text)
```
