```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ['OPENAI_API_KEY'] = os.getenv('OPENAI_API_KEY')
##Langsmith tracking
os.environ["LANGCHAIN_API_KEY"] = os.getenv("LANGCHAIN_API_KEY")
os.environ["LANGCHAIN_PROJECT"] = os.getenv("LANGCHAIN_PROJECT")
os.environ["LANGCHAIN_TRACING_V2"] = "true"
```

```python
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(model = "gpt-4o")
print(llm)
```

```python
#Input and get response from llm
result = llm.invoke("Is Publicis Sapient a good company to work in field of Machine Learning?")
```

```python
print(result)
```

```python
### ChatPrompt Template
from langchain.prompts import ChatPromptTemplate

input = "Is Publicis Sapient a good company to work in field of Machine Learning?"

prompt = ChatPromptTemplate.from_messages(
    [
        ("system" , "You are a helpful assistant that provides information about companies."),
        ("user", "{input}")
    ])

prompt
```

```python
#Creating a chain 

chain = prompt | llm #Combines the prompt and the llm
#Invoke the chain with input
result = chain.invoke({"input": input})
```

```python
print(result)
print(type(result))
```

```python
#Use output parser to structure the response

from langchain_core.output_parsers import StrOutputParser
output_parser = StrOutputParser()

#Craete a new chain 
chain = prompt | llm | output_parser
```

```python
response = chain.invoke({"input": "Compare Publicis Sapient and ZS Associates in terms of their Machine Learning capabilities."})
print(response)
```
